##### 1. Energy Consumption per Grid Area in Denmark

https://www.energidataservice.dk/tso-electricity/ConsumptionPerGridarea

```python
energy = response = requests.get(url='https://api.energidataservice.dk/dataset/ConsumptionPerGridarea?limit=10000')
```

```python
energy.status_code # Check if ok. 200 means all good
```

```python
#energy.content
#energy.json()  Energy is a json
```

```python
energy_data = pd.DataFrame(energy.json())
energy_data.head()
# As wee see al information is in records column. We need it to extract it an make it a df
```

```python
energy_data['records']
dict(energy_data['records'])
pd.DataFrame(dict(energy_data['records'])).T
```

##### 2. Crypto Exchange

https://docs.coincap.io/

```python
crypto = requests.get("https://api.coincap.io/v2/assets")
```

```python
crypto.status_code
```

```python
#crypto.content
```

```python
crypto_data = pd.DataFrame(crypto.json())
crypto_data.head() # Again. All information is in data column. We need to extract it
```
